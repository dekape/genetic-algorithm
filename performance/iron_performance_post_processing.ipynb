```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from matplotlib import rcParams
rcParams.update({'figure.autolayout': True})
```

```python
# CROSS-OVER

# 5units
filename = "stats5_pcrossing.csv"
header = 0
data = pd.read_csv(filename, header=header)

iter_count = data["iter_count"]
p_crossing = data["p_crossing"]

avg_itr_count_pcross5 = []
avg_pcross5 = p_crossing[::10]
for i in range (0, avg_pcross5.size):
    avg_itr_count_pcross5.append(np.mean(iter_count[i*10:(i+1)*10]))
    
# 10 units
filename = "stats10_pcrossing.csv"
data = pd.read_csv(filename, header=header)

iter_count = data["iter_count"]
p_crossing = data["p_crossing"]

avg_itr_count_pcross10 = []
avg_pcross10 = p_crossing[::10]
for i in range (0, avg_pcross10.size):
    avg_itr_count_pcross10.append(np.mean(iter_count[i*10:(i+1)*10]))
```

```python
# MUTATION
filename = "stats5_pmutation.csv"
header = 0
data = pd.read_csv(filename, header=header)

iter_count = data["iter_count"]
p_mutation = data["p_mutation"]

avg_itr_count_pmutate5 = []
avg_pmutation5 = p_mutation[::10]
for i in range (0, avg_pmutation5.size):
    avg_itr_count_pmutate5.append(np.mean(iter_count[i*10:(i+1)*10]))
    
filename = "stats10_pmutation.csv"
header = 0
data = pd.read_csv(filename, header=header)

iter_count = data["iter_count"]
p_mutation = data["p_mutation"]

avg_itr_count_pmutate10 = []
avg_pmutation10 = p_mutation[::10]
for i in range (0, avg_pmutation10.size):
    avg_itr_count_pmutate10.append(np.mean(iter_count[i*10:(i+1)*10]))
```

```python
# NUMBER OF CIRCUITS
filename = "stats5_nocircuit.csv"
header = 0
data = pd.read_csv(filename, header=header)

iter_count = data["iter_count"]
no_circuits = data["no_circuits"]

avg_itr_count_nocirc5 = []
avg_nocirc5 = no_circuits[::10]
for i in range (0, avg_nocirc5.size):
    avg_itr_count_nocirc5.append(np.mean(iter_count[i*10:(i+1)*10]))
    
filename = "stats10_nocircuit.csv"
data = pd.read_csv(filename, header=header)

iter_count = data["iter_count"]
no_circuits = data["no_circuits"]

avg_itr_count_nocirc10 = []
avg_nocirc10 = no_circuits[::10]
for i in range (0, avg_nocirc10.size):
    avg_itr_count_nocirc10.append(np.mean(iter_count[i*10:(i+1)*10]))
```

```python
# PLOT

print("ITERATIONS AVERAGED OVER 10 EXECUTIONS")
print("Fitness (5 units)         : ", 24.8163)
print("Fitness (10 units)        : ", 165.753)
print("Max iter                  : ", 5000)
print("Iter convergence          : ", 200)
print("Standar p_mutation        : ", 0.01)
print("Standar p_crossing        : ", 0.9)
print("Standar number of circuits: ", 100)

fig, ax = plt.subplots(1,3,figsize=(15, 7))
fig.tight_layout(w_pad = 2)
ax[0].plot(avg_pcross5, avg_itr_count_pcross5, ".-", label=" 5 Units")
ax[0].plot(avg_pcross10, avg_itr_count_pcross10, ".-", label=" 10 Units")
ax[0].set_title("Performance vs Crossing Probability")
ax[0].set_xlabel("Crossing Probability")


ax[1].plot(avg_pmutation5, avg_itr_count_pmutate5, ".-", label=" 5 Units")
ax[1].plot(avg_pmutation10, avg_itr_count_pmutate10, ".-", label=" 10 Units" )
ax[1].set_title("Performance vs Mutation Probability")
ax[1].set_xlabel("Mutation Probability")

ax[2].plot(avg_nocirc5, avg_itr_count_nocirc5, ".-", label=" 5 Units")
ax[2].plot(avg_nocirc10, avg_itr_count_nocirc10, ".-", label=" 10 Units")
ax[2].set_title("Performance vs Number of Circuits")
ax[2].set_xlabel("Number of Circuits")

for i in range(0,3):
    ax[i].grid();
    ax[i].legend(loc = "best")
    ax[i].set_ylabel(" Average Number of Iterations for Convergence")
plt.show()
fig.savefig("plots.png")

print("Notes:\n crossing 10 units: first point might be an outlier \n circuits 10 units: first point might be an outlier \n hard to do performance because of probabilistic approach, if time would do over more executions instead of 10")
```
